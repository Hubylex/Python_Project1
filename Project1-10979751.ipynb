### Getting rid of the null values by replacing them in each column by the MEAN VALUES
The mean was used to replace the null values because the data is normally distributed. The mean is a good choice when the data is evenly distributed around the center and there are no extreme values pulling the mean in one direction or another.And in this case of data it is.

- Calculating mean values for each column

```python

mean_c1 = data['ID'].mean().round(2)
mean_c2 = data['fixed acidity'].mean().round(2)
mean_c3 = data['volatile acidity'].mean().round(2)
mean_c4 = data['citric acid'].mean().round(2)
mean_c5 = data['residual sugar'].mean().round(2)
mean_c6 = data['chlorides'].mean().round(2)
mean_c7 = data['free sulfur dioxide'].mean().round(2)
mean_c8 = data['total sulfur dioxide'].mean().round(2)
mean_c9 = data['density'].mean().round(2)
mean_c10 = data['pH'].mean().round(2)
mean_c11 = data['sulphates'].mean().round(2)
mean_c12 = data['alcohol'].mean().round(2)
mean_c13 = data['quality'].mean().round(2)
```

- Replacing the null values in each column by the MEAN VALUES

```python
data['ID'] = data['ID'].fillna(mean_c1)
data['fixed acidity'] = data['fixed acidity'].fillna(mean_c2)
data['volatile acidity'] = data['volatile acidity'].fillna(mean_c3)
data['citric acid'] = data['citric acid'].fillna(mean_c4)
data['residual sugar'] = data['residual sugar'].fillna(mean_c5)
data['chlorides'] = data['chlorides'].fillna(mean_c6)
data['free sulfur dioxide'] = data['free sulfur dioxide'].fillna(mean_c7)
data['total sulfur dioxide'] = data['total sulfur dioxide'].fillna(mean_c8)
data['density'] = data['density'].fillna(mean_c9)
data['pH'] = data['pH'].fillna(mean_c10)
data['sulphates'] = data['sulphates'].fillna(mean_c11)
data['alcohol'] = data['alcohol'].fillna(mean_c12)
data['quality'] = data['quality'].fillna(mean_c13)
```
